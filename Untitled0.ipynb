```python
# Questão 16

# List Comprehension: [expressão for item in iterável if condição]
cubos_pares = [num**3 for num in range(11) if num % 2 == 0]

# Imprimindo a lista resultante
print(f"Cubo dos números pares de 0 a 10: {cubos_pares}")
```

```python
# Questão 17

# Definindo a função
def saudacao_personalizada(nome):
    """
    Imprime uma saudação personalizada para o usuário.
    """
    print(f"Seja muito bem-vindo(a) ao mundo dos dados, {nome}!")

# Testando a função
saudacao_personalizada("Maria")
saudacao_personalizada("João")
```

```python
# Questão 18

# Definindo a função com retorno
def calcula_desconto(valor_produto, porcentagem_desconto):
    """
    Calcula e retorna o valor final do produto após aplicar o desconto.
    """
    valor_descontado = valor_produto * (porcentagem_desconto / 100)
    valor_final = valor_produto - valor_descontado
    return valor_final

# Testando a função
preco_original = 100.00
desconto = 15  # 15%

resultado = calcula_desconto(preco_original, desconto)

# Imprimindo o resultado formatado
print(f"O valor original era R$ {preco_original:.2f}")
print(f"Com {desconto}% de desconto, o valor final é R$ {resultado:.2f}")
```

```python
# Questão 19

# Definindo a função de classificação
def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"

# Chamando a função e armazenando o retorno em uma variável
minha_nota = 6.5
status = classifica_nota(minha_nota)

# Imprimindo o resultado
print(f"Com a nota {minha_nota}, o aluno está: {status}")
```

```python
# Questão 20

# Lista de números mistos
numeros = [10, -5, 0, 22, -8, 0, -3, 15, 42, 0, -1]

# Inicializando os contadores
positivos = 0
negativos = 0
zeros = 0

# Laço para percorrer a lista
for num in numeros:
    if num > 0:
        positivos += 1
    elif num < 0:
        negativos += 1
    else:
        zeros += 1

# Imprimindo o resumo final
print("--- Resumo da Contagem ---")
print(f"Positivos: {positivos}")
print(f"Negativos: {negativos}")
print(f"Zeros:     {zeros}")
```
